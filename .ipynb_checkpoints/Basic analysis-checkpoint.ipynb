# Data preparation

```python
df = pd.read_csv("data/train2.tsv", sep="\t", names=[
    "id",
    "json_id",
    "label",
    "statement",
    "subject",
    "speaker",
    "job_title",
    "state",
    "party_affiliation",
    "barely_true_counts",
    "false_counts",
    "half_true_counts",
    "mostly_true_counts",
    "pants_on_fire_counts",
    "context",
    "justification"
])
```

```python
df.head(2)
```

```python
df.info()
```

```python
df[["statement", "justification"]] = df[["statement", "justification"]].astype(str)
```

```python
df.info()
```

## Add statement and justification length information

```python
df["statement_len"] = df["statement"].apply(len)
df["justification_len"] = df["justification"].apply(len)
```

```python
df = df.drop(["id", "json_id"], axis=1)
```

```python
df.head(2)
```

## Add punctuations count per 100 characters

```python
string.punctuation
```

```python
df["punctuations"] = df["statement"].apply(lambda s: sum(map(s.count, string.punctuation)))
```
